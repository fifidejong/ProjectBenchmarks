```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library

%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

df_ = pd.read_excel('Book1.xlsx', index_col=0)

print('Data downloaded and read into a dataframe!')
```

```python
df_.head()
```

```python
df_select        = df_[['AFIPSKY HCU','OMSK AOPU','GCC','SIBUR','Category']]
df_select        = df_select.sort_values(['SIBUR'],ascending=True, axis=0)

df_select_dispct        = df_[['Distr. Phase-1','Distr. Phase-2','Distr. Phase-3','Distr. Phase-4','Category']]

df_select_Eng    = df_select[df_select['Category']==1]
df_select_Eng    = df_select_Eng[['AFIPSKY HCU','OMSK AOPU','GCC','SIBUR']]

df_select_NonEng = df_select[df_select['Category']==2]
df_select_NonEng = df_select_NonEng[['AFIPSKY HCU','OMSK AOPU','GCC','SIBUR']]

df_select_Eng_dispct    = df_select_dispct[df_select_dispct['Category']==1]
df_select_Eng_dispct    = df_select_Eng_dispct[['Distr. Phase-1','Distr. Phase-2','Distr. Phase-3','Distr. Phase-4']]*100

df_select_NonEng_dispct = df_select_dispct[df_select_dispct['Category']==2]
df_select_NonEng_dispct = df_select_NonEng_dispct[['Distr. Phase-1','Distr. Phase-2','Distr. Phase-3','Distr. Phase-4']]*100
```

```python
# generate plot

columns          = ['AFIPSKY HCU','OMSK AOPU','GCC','SIBUR']
colors           = ['#f0ad4e', '#5bc0de', '#d9534f','#5cb85c' ]

df_select_Eng.plot(kind='barh', figsize=(15, 15), width = 0.8, color = colors, edgecolor=None, fontsize=14)
plt.xlabel('Total manhour',fontsize=16)
plt.ylabel('Engineering Discipline',fontsize=16)
plt.title('Manhour Distribution per Engineering Discipline',fontsize=16)
plt.legend(labels= columns, fontsize=16)

# get the axis
ax = plt.gca()

#ax.set_xticks(fontsize=14)
for spine in ax.spines.values():
   spine.set_visible(False)
#ax.set_yticks([])
    
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    mhr_val = p.get_width()
        
    top_label = str("{:,}".format(mhr_val))
    ax.annotate(top_label, xy=(x+ width + 5, y+0.05 ), color='black', fontsize=14)

plt.show()
```

```python
colors_list = ['#b42830', '#c39bc8', '#ffa723', '#00b092', '#568bd2', '#e2947b', '#c7ced5']
explode_list = [0, 0, 0, 0, 0, 0, 0] # ratio for each continent with which to offset each wedge.

df_select_Eng['AFIPSKY HCU'].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,         # turn off labels on pie chart
                            pctdistance=1.12,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                            colors=colors_list,  # add custom colors
                            explode=explode_list, # 'explode' lowest 3 continents
                            fontsize = 14      
                            )

# scale the title up by 12% to match pctdistance
plt.title('Engineering manhour distribution of Afipsky HCU', y=1.12, fontsize=14) 
plt.ylabel(None)
plt.axis('equal') 

# add legend
plt.legend(labels=df_select_Eng.index, loc='upper left', fontsize=14) 

plt.show()
```
